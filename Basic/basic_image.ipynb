```python
cap = cv.VideoCapture('../Resource/Videos/sample_video.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Perform inference on the frame using the model
    results = model.track(frame, stream=True)

    # Iterate through each result (predicted box)
    for result in results:
        class_names = result.names  # Get the class names

        # Iterating over each box detected
        for box in result.boxes:
            if box.conf[0] > 0.4:  # Confidence threshold
                # Get coordinates for the bounding box
                x1, y1, x2, y2 = box.xyxy[0]

                # Convert coordinates to integers
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

                # Get the class name
                class_name = class_names[int(box.cls[0])]

                # Draw the bounding box
                cv.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                # cv.putText(frame, f"{class_name} {box.conf[0]:.2f}", (x1, y1 - 10), 
                #            cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the frame with bounding boxes
    cv.imshow("YOLO Video Detection", frame)

    # Wait for 1 ms, exit if 'q' is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close the display window
cap.release()
cv.destroyAllWindows()
```
